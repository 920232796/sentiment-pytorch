````
import re
pos_data = ""
target_pos = []
with open("./pos.txt", "r") as f1:
  pos_data = f1.read().replace("，", " ").replace("。", " ").replace("！", " ").replace("？", " ").replace(",", " ").replace(" ", "").replace("(", "").split("\n")
  print(pos_data[:10])
  for index, each_sentence in enumerate(pos_data):
    pos_data[index] = re.sub("[A-Za-z0-9\!\%\[\]\,\。\(\)\-\？\?]", "", each_sentence)
pos_data1 = [[each for each in sentence] for sentence in pos_data]
print(pos_data1[:5])
target_pos = [1 for i in range(len(pos_data))]
print(len(target_pos))

neg_data = ""
target_neg = []
with open("./neg.txt", "r") as f2:
  neg_data = f2.read().replace("，", " ").replace("。", " ").replace("！", " ").replace("？", " ").replace(",", " ").replace(" ", "").split("\n")
  
  for index, each_sentence in enumerate(neg_data):
    neg_data[index] = re.sub("[A-Za-z0-9\!\%\[\]\,\。\(\)\-\？\?]", "", each_sentence)
  print(neg_data[:10])
neg_data1 = [[each for each in sentence] for sentence in neg_data]
target_neg = [0 for i in range(len(neg_data))]
print(target_neg)

print(len(neg_data1))
print(len(target_neg))
````

````
target_data = target_pos + target_neg 
print(target_data[:10])
data = pos_data1 + neg_data1
print(data[:2])
print(len(data))

##  现在得到了 data，然后还有data中每个句子对应的target是0 还是 1 

## 清洗一下数据～

new_target = target_data
new_data = []

max_length = 30
for index, sentence in enumerate(data):
  if (len(sentence) < max_length):
    sentence += ["pad" for i in range(max_length - len(sentence))] ## 太短的句子要补 pad 补成一样长！
    new_data.append(sentence)
    
  else :
    new_data.append(sentence[:max_length])

# senteces = [[char2ix[word] for sentence in new_data] for word in sentence]
print(new_data[:5])
print(new_target[:5])

print(len(new_data))
````

````
## 开始构建indice，也就是每个字都对应一个数字！
ix2char = list(set([c for sentence in new_data for c in sentence]))
ix2char.append("pad") ## 把填充的项放到最后面把
print(ix2char)
char2ix = {key:value for value, key in enumerate(ix2char)}
print(char2ix)
vocab_size = len(ix2char)
print(vocab_size)
````

````
##  现在可以将 字 转为对应的indice了 然后以后输入到embeeding层里面去

data_indice = [[char2ix[word] for word in sentence] for sentence in new_data]
print (data_indice)
print(len(data_indice))
````

````
## 到现在句子也补全了，太长的也裁剪了，都一样长了。
import torch
## 先打乱一下顺序

import numpy as np

shuffle_ix = np.random.permutation(np.arange(len(data_indice)))

train_data = np.array(data_indice)[shuffle_ix]
train_label = np.array(new_target)[shuffle_ix]


## 写一个取batch size 数据的函数

def get_data_iter(data_X, data_Y, batch_size, device):
  batch_len = len(data_X) // batch_size
  data_X = torch.from_numpy(data_X)
  # print(data_X.shape)
 
  data_Y = torch.from_numpy(data_Y)
  
  for each_batch in range(batch_len):

    X = data_X[each_batch * batch_size:(1+each_batch)*batch_size]
    Y = data_Y[each_batch * batch_size:(1+each_batch)*batch_size]
    X.to(device)
   
    X = X.transpose(0, 1)
   
    Y.to(device)

    yield X, Y


data_iter = get_data_iter(train_data, train_label, 5, "cpu")

for X, Y in data_iter:
  print(X)
  print(Y)
  break
````

````
## 构建模型～

## 建立模型
from torch import nn
import d2lzh_pytorch as d2l

import time
word_vec_dim = 64
num_hiddens = 32 ## rnn里面 运算的时候 比如 input.dot(H) Hshape为:(word_vec_dim, hidden_num)
max_length = train_data.shape[1] ## 也就是时间步

embedding_layer = nn.Embedding(len(ix2char), word_vec_dim)
device = "cpu"

rnn_layer = nn.RNN(input_size=word_vec_dim, hidden_size=num_hiddens)

class RNNModel(nn.Module):

  def __init__(self, embedding_layer, rnn_layer, max_length):
    super(RNNModel, self).__init__()
    self.embedding = embedding_layer
    self.rnn = rnn_layer
    self.hidden_size = rnn_layer.hidden_size 
    self.max_length = max_length
    self.dense = nn.Linear(self.hidden_size * max_length, 2) ## 全联接层，最后输出用 输出两类！  这里注意阿！ 前面是得* max_length的！！因为所有时间步的y连起来
    self.state = None 
  
  def forward(self, inputs, state):

    inputs = self.embedding(inputs)

    Y, self.state = self.rnn(inputs, state)

    ##全联接层 ，操作就是 output = Y.dot(Wy) + b  Wy.shape:(hidden_size, outputs_num)
    output = self.dense(Y.view(inputs.shape[1], -1))## 
    ## output shape为：(num_steps * batch_size, vocab_size)
    return output, self.state


rnn_model = RNNModel(embedding_layer, rnn_layer, 30).to(device)
# state = None
# X_tmp = torch.randn(30, 2) + 100
# X_tmp = X_tmp.long()
# out, state = rnn_model(X_tmp, state)
# out = embedding_layer(X_tmp)



# print(out.shape)
````

````
## 训练函数

def train_model(model, data, label, num_hiddens, num_epochs, max_length, lr, clipping_theta, batch_size):
  loss = nn.CrossEntropyLoss()
  optimizer = torch.optim.Adam(model.parameters(), lr)
  device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
  state = None
  model.to(device)

  for epoch in range(num_epochs):
    l_sum, n, start = 0.0, 0, time.time()
    data_iter = get_data_iter(data, label, batch_size, device)
    # print(epoch)

    for X, y in data_iter:

      if state is not None:
        if isinstance (state, tuple): # LSTM, state:(h, c)  
          state = (state[0].detach(), state[1].detach())
        else:   
          state = state.detach()
      (output, state) = model(X, state)

      l = loss(output, y.long())
      optimizer.zero_grad()##梯度清零
      l.backward()

      d2l.grad_clipping(model.parameters(), clipping_theta, device)
      optimizer.step()
      l_sum += l.item() * batch_size ## shape[0] 是batch_size * num_step 因为求loss的时候 除以N了 


    print('epoch %d, time %.2f sec, loss %.8f' % (
            epoch + 1, time.time() - start, l_sum))
````

````
num_epochs, batch_size, lr, clipping_theta = 70, 32, 1e-3, 1e-2 # 注意这里的学习率设置
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
train_model(rnn_model, train_data, train_label, num_hiddens, num_epochs,
                            30, lr, clipping_theta, batch_size)


torch.save(rnn_model, "./model_embedding2.pkl")
````
